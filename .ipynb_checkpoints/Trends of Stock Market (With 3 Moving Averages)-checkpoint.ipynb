```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
plt.style.use('fivethirtyeight')
```

```python
data = pd.read_csv('AMD_Long.csv')
data["Close/Last"] = data["Close/Last"].str.replace("$","")
data["Open"] = data["Open"].str.replace("$","")
data["High"] = data["High"].str.replace("$","")
data["Low"] = data["Low"].str.replace("$","")

column_names = ['Close/Last', 'Open', 'High', 'Low']
for i in column_names: 
    data[i] = pd.to_numeric(data[i], downcast = "float")
    
data = data.reindex(index=data.index[::-1])
data
```

```python
# set the date to be the index

data = data.set_index(pd.DatetimeIndex(data['Date'].values))
data = data.rename(columns = {'Close/Last' : 'Close'})
```

```python
data
```

```python
# plotting the close price
plt.figure(figsize = (12.2, 4.5))
plt.title('AMD (Advanced Micro Devices)')
plt.xlabel('Date', fontsize = 18)
plt.ylabel('Close Price ($)', fontsize = 18)
plt.plot(data['Close'])
plt.show()
```

### 1. Using EMA (Exponential Moving Averages)

```python
# short/fast Exponential Moving Average
Short_EMA = data.Close.ewm(span = 20, adjust = False).mean()

# middle/medium Exponential Moving Average
Middle_EMA = data.Close.ewm(span = 50, adjust = False).mean()

# long/slow Exponential Moving Average
Long_EMA = data.Close.ewm(span = 60, adjust = False).mean()
```

```python
data['Short/Fast EMA'] = Short_EMA
data['Middle/Medium EMA'] = Middle_EMA
data['Long/Slow EMA'] = Long_EMA
data
```

```python
# plotting everything
plt.figure(figsize = (12.2, 4.5))
plt.title('AMD (Advanced Micro Devices)')
plt.xlabel('Date', fontsize = 18)
plt.ylabel('Close Price($)')
plt.plot(data['Close'])
plt.plot(data['Short/Fast EMA'], color = 'black')
plt.plot(data['Middle/Medium EMA'], color = 'red')
plt.plot(data['Long/Slow EMA'], color = 'yellow')
plt.show()
```

```python
# 2nd way to plot

close_df = data.loc[:, ['Close']]
close_df['Short/Fast EMA'] = Short_EMA
close_df['Middle/Medium EMA'] = Middle_EMA
close_df['Long/Slow EMA'] = Long_EMA
close_df
```

```python
close_df.plot()
plt.show
```

```python
# Creating the function to buy and sell the stock 
def buying_selling(data):
    
    buy = []
    sell = []
    
    ammount = 10000.0
    hold_stocks = 0.0
    
    buyandHoldStocks = ammount / data['Close'][0]
    buy.append(np.nan)
    sell.append(np.nan)
    
    for i in range(1, len(close_df)):
        # it means the trend is going up, that's why it makes sense to buy 
        if close_df['Short/Fast EMA'][i] > close_df['Middle/Medium EMA'][i] and close_df['Short/Fast EMA'][i-1] < close_df['Middle/Medium EMA'][i-1]:            
            hold_stocks = ammount / data['Close'][i]
            buy.append(data['Close'][i])
            sell.append(np.nan)
        # it means the trend is going down, so we sell 
        elif close_df['Short/Fast EMA'][i] < close_df['Middle/Medium EMA'][i] and close_df['Short/Fast EMA'][i-1] > close_df['Middle/Medium EMA'][i-1]:
            hold_stocks = -(ammount / data['Close'][i])
            sell.append(data['Close'][i])
            buy.append(np.nan)
        else:
            buy.append(np.nan)
            sell.append(np.nan)
        if hold_stocks < 0:
                ammount += (data['Close'][i-1] - data['Close'][i]) * (hold_stocks * -1.0)
        if hold_stocks > 0:
            ammount = hold_stocks * data['Close'][i]
        
    bh =  buyandHoldStocks * data['Close'][len(close_df) - 1]   
    print("vermoegen buy and hold " , str(bh))
    print("profit trading" , str(ammount))
    return(buy, sell)
```

```python
close_df['Buy'] = buying_selling(close_df)[0]
close_df['Sell'] = buying_selling(close_df)[1]
close_df
```

```python
plt.figure(figsize = (12.2, 4.5))
plt.title('AMD (Advanced Micro Devices)')
plt.xlabel('Date', fontsize = 18)
plt.ylabel('Close Price($)')
plt.plot(data['Close'], color = 'blue', alpha = 0.3) 
plt.plot(data['Short/Fast EMA'], color = 'black', alpha = 0.25)
plt.plot(data['Middle/Medium EMA'], color = 'red', alpha = 0.25)
#plt.plot(data['Long/Slow EMA'], color = 'yellow', alpha = 0.25)
plt.scatter(close_df.index, close_df['Buy'], color = 'green', marker = '^', alpha = 1)
plt.scatter(close_df.index, close_df['Sell'], color = 'red', marker = 'v', alpha = 1)
plt.show()
```
